# MAIN EXPERIMENT :: BOX D

```python
BoxD = pd.read_excel("Lab1 Python.xlsx",sheet_name="BoxD")
BoxD = BoxD.drop(labels=[20], axis=0)
BoxD['|Z|'] = (BoxD['Resistance Ohms (Ω)'] * BoxD['Amp CH1 (V)'] / BoxD['Amp CH2 (V)']).abs()
BoxD['|Z|^2'] = BoxD['|Z|']**2
BoxD['w'] = BoxD['Average Frequency (Hz)']*2*np.pi
BoxD['1/w'] = 1/BoxD['w']
BoxD['w^2'] = BoxD['w']**2
BoxD['1/w^2'] = 1/BoxD['w^2']
```

```python
BoxD
```

### ERROR PROPAGATION

```python
f = list(map(lambda x: 1.0 if (x>10) else 0.1, BoxD['Resistance Ohms (Ω)']))
BoxD['uncert_VCH1'] = 0.03*BoxD['Amp CH1 (V)']
BoxD['uncert_VCH2'] = 0.03*BoxD['Amp CH2 (V)']
BoxD['uncert_R'] = f
BoxD['sigma_Y'] = np.sqrt(((BoxD['Resistance Ohms (Ω)']*BoxD['Amp CH1 (V)']*BoxD['uncert_VCH2'])**2/BoxD['Amp CH2 (V)']**4)+((BoxD['Resistance Ohms (Ω)']*BoxD['uncert_VCH1'])**2/BoxD['Amp CH2 (V)']**2)+((BoxD['Amp CH1 (V)']*BoxD['uncert_R'])**2/BoxD['Amp CH2 (V)']**2))
BoxD['sigma_Y^2'] = np.sqrt(((BoxD['Resistance Ohms (Ω)']**2 *BoxD['Amp CH1 (V)']**2 *BoxD['uncert_VCH2'])**2/BoxD['Amp CH2 (V)']**6)+((BoxD['Resistance Ohms (Ω)']**2 *BoxD['uncert_VCH1'])**2/BoxD['Amp CH2 (V)']**4)+((BoxD['Amp CH1 (V)']**2 *BoxD['uncert_R'])**2/BoxD['Amp CH2 (V)']**4))
```

### PLOT BOXD

```python
l = 2
k = 17
m2, b2 = np.polyfit(BoxD['1/w^2'][l:k],BoxD['|Z|^2'][l:k], 1)
Chi_Sqrd = np.sum((BoxD['|Z|^2'][l:k]-(BoxD['1/w^2'][l:k]*m2+b2))**2 / (BoxD['sigma_Y^2'][l:k]**2))
N = len(BoxD['1/w^2'][l:k])
sigma_y = np.sqrt((1/(N-2))*np.sum((BoxD['|Z|^2'][l:k]-(BoxD['1/w^2'][l:k]*m2+b2))**2))
delta = N*np.sum(BoxD['1/w^2'][l:k]**2)-(np.sum(BoxD['1/w^2'][l:k]))**2
uncert_B2 = sigma_y*np.sqrt(np.sum(BoxD['1/w^2'][l:k]**2)/delta)
uncert_M2 = sigma_y*np.sqrt(N/delta)
# print(uncert_B2, uncert_M2)
print("sigmaY =",sigma_y, "delta =", delta)
ax2 = plt.figure().add_subplot(111)
ax2.scatter(BoxD['1/w^2'][l:k], BoxD['|Z|^2'][l:k])
ax2.errorbar(BoxD['1/w^2'][l:k],BoxD['|Z|^2'][l:k], yerr=BoxD['sigma_Y^2'][l:k], capsize=4, ecolor='red', elinewidth=1, markeredgewidth=1, linestyle='None')
plt.title('Impedance $\|Z\|^2$ vs. Inverse Angluar Frequencies $\dfrac{1}{\omega^2}$', fontsize=14)
plt.xlabel('$\dfrac{1}{\omega^2}$ / $\dfrac{1}{Hz^2}$', fontsize=12)
plt.ylabel('$\|Z\|^2$ / $\Omega^2$',fontsize=12)

# ax2.scatter(x=BoxD['w^2'],y=(BoxD['w^2']*m2+b2), label='Predicted |Y|^2', color='orange')
# ax2.set_yscale('log')
# ax2.set_xscale('log')

plt.plot(BoxD['1/w^2'][l:k],m2*BoxD['1/w^2'][l:k]+b2, '-',color='skyblue', label='y=({:.2e} $\pm$ {:.2e})x +\n{:.2e} $\pm$ {:.2e}'.format(m2,uncert_M2,b2,uncert_B2))
plt.plot([], [], ' ', label='$\chi^2 = ${:n}, should be $\leq$ {:n}'.format(Chi_Sqrd, len(BoxD['w^2'][l:k])))
ax2.plot(BoxD['1/w^2'][l:k],(BoxD['1/w^2'][l:k]*(m2+uncert_M2)+b2+uncert_B2), '--', alpha=0.2, label='Upper Bound')
ax2.plot(BoxD['1/w^2'][l:k],(BoxD['1/w^2'][l:k]*(m2-uncert_M2)+b2-uncert_B2), '--', alpha=0.2, label='Lower Bound')

print("Chi^2 is", Chi_Sqrd, "when it should be less than or equal to", len(BoxD['w^2'][l:k]))
R_theory = np.sqrt(b2)
C_theory = np.sqrt(1/m2)*1e9
plt.plot([], [], ' ', label='Capcitance: {:.1f} $\pm$ {:.1f} nF\nResistance: {:.1f} $\pm$ {:.1f} $\Omega$'.format(C_theory, (uncert_M2/m2)*C_theory, R_theory, np.sqrt(uncert_B2)))
print(str(R_theory)+" Ohms",str(C_theory)+"nF")
ax2.legend(loc='upper left', fontsize=14)
```
